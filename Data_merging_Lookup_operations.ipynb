**Import Libraries:**

What are the library needed to import the library files

```python
import pandas as pd
import numpy as np
```

**Task 1: Data Preparation and Missing Value Handling**

**Create Dataframe**

Create the dataframe of students_data,enrollments_data,scores_data.These three datas are creation of dataframe.

```python
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}

enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}

scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}

students = pd.DataFrame(students_data)
enrollments = pd.DataFrame(enrollments_data)
scores = pd.DataFrame(scores_data)


```

**Display Original Students DataFrame**

Once creation completed,after the dataframe should displayed at output page.

```python
print("Original Students DataFrame:")
print(students)
```

**Null Value Count and Percentage**

All null value count and percentage also check after that provide the values are  displayed on output page.

```python
print("\nNull Value Analysis")

for column in students.columns:
  null_count = students[column].isnull().sum()
  null_percentage = (null_count / len(students)) * 100
  print(f"Column: {column}, Nulls:{null_count} ({null_percentage:.2f}%)")
```

**Handle Missing Values**

Fill the missing values at the city column,Null value change into Unknown.

```python
students['city'].fillna('Unknown', inplace=True)
print("\nStudents DataFrame after handling missing values:")
print(students)
```

**Drop rows where name is missing**

Name-column have null value,it will drop or remove the null value of row.(103 student id-rows are dropped.)

```python
students_clean = students.dropna(subset=['name'])
```

**Display Cleaned DataFrame**

Once above step complete,after the Displayed the cleaned Dataframe.



```python
print("\nCleaned Students DataFrame:")
print(students_clean)
```

**Task 2:Multiple Join Operations**

**1.Inner Join**

A data merging technique that combines two DataFrames based on a common key or column, returning only the rows that have matching values in both tables.

```python
inner_join = pd.merge(students_clean, enrollments, on='student_id', how='inner')

print("\nInner Join Result:")
print(inner_join)
print("Total rows:", len(inner_join))
```

2.Left Join

Combines two DataFrames and keeps all rows from the left DataFrame, along with any matching rows from the right DataFrame.

```python
left_join = pd.merge(students_clean, enrollments, on='student_id', how='left')

print("\nLeft Join Result:")
print(left_join)
print("Total rows:", len(left_join))
```

**Right Join**

Keeps all records from the right DataFrame (the lookup table) and matches them with records from the left DataFrame

```python
right_join= pd.merge(students_clean, enrollments, on='student_id', how='right')

print("\nRight Join Result:")
print(right_join)
print("Total rows:", len(right_join))
```

**Full Outer Join**

Merges two DataFrames (or a lookup table and a main dataset) by including all records from both, matching them wherever possible.


```python
outer_join= pd.merge(students_clean, enrollments, on='student_id', how='outer')

print("\nOuter Join Result:")
print(outer_join)
print("Total rows:", len(outer_join))
```

**Missing rows to check is null both of name and course_name.**

```python
missing_rows= outer_join[
    outer_join['name'].isnull() | outer_join['course_name'].isnull()]

print("\nMissing Rows:")
print(missing_rows)
```

**Indicator Column**

It is a special column added to a merged DataFrame by setting indicator=True in the pd.merge() function.t creates a column, usually named _merge, that identifies if a row exists in the left table, right table, or both, providing three values: left_only, right_only, or both.

```python
outer_join_indicator = pd.merge(
    students_clean,
    enrollments,
    on='student_id',
    how='outer',
    indicator=True
)

print(outer_join_indicator['_merge'].value_counts())
```

**Task 3:Lookup Operation**

**Create Dictionary**


Score_dict new variable created.Student_id scores and exam_score also zip and store the score_dict file.

```python
score_dict = dict(zip(scores['student_id'], scores['exam_score']))
print(score_dict)
```

**Add Scores using map()**

Add the score of different column added.(mention column name student_id mapped with score_dict.

```python
students_clean['exam_score'] = students_clean['student_id'].map(score_dict)

print(students_clean[['student_id','name','exam_score']])
```

**Performance Comparison(merge)**

pd.merge() in pandas for lookup tables involves evaluating speed differences between column-based merging (slower) and index-based merging (faster).

```python
score_merge= pd.merge(
    students_clean,
    scores,
    on='student_id',
    how='left'
)

# Access 'exam_score_x' from students_clean or 'exam_score_y' from scores
print(score_merge[['student_id','name','exam_score_x', 'exam_score_y']])
```

**Task 3: Automation Function**

It acts as a powerful, SQL-style, or VLOOKUP-like tool to automatically align and join datasets based on shared columns or indices

```python
def auto_merge(df1, df2, join_type, key_column):

  merged_df = pd.merge(df1, df2, on=key_column, how=join_type)

  return {
      'result_df': merged_df,
      'row_count': len(merged_df),
      'join_type': join_type,

  }
```

**Test the Function**

**Inner Join Test**

Check the inner join test of function.

```python
result1 = auto_merge(students_clean, enrollments, 'inner', 'student_id')

print("Join type:", result1['join_type'])
print("Rows in Result:", result1['row_count'])
print(result1['result_df'].head())
```

**Left Join Test**
Check the Left join test of function.

```python
result2 = auto_merge(students_clean, enrollments, 'left', 'student_id')

print("Join type:", result1['join_type'])
print("Rows in Result:", result1['row_count'])
print(result2['result_df'].head())
```
