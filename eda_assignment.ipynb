**Step 1 — Generate the Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

n = 200

data = {
    "order_id": range(1001, 1001 + n),
    "city": np.random.choice(["Mumbai", "Delhi", "Bangalore", "Chennai"], size=n),
    "category": np.random.choice(["Electronics", "Clothing", "Groceries", "Furniture"], size=n),
    "order_value": np.random.randint(200, 5000, size=n).astype(float),
    "delivery_days": np.random.randint(1, 15, size=n).astype(float),
    "rating": np.random.choice([1, 2, 3, 4, 5, None], size=n)
}

missing_indices_order = np.random.choice(n, size=15, replace=False)
missing_indices_delivery = np.random.choice(n, size=10, replace=False)

data["order_value"][missing_indices_order] = np.nan
data["delivery_days"][missing_indices_delivery] = np.nan

data["order_value"][5] = 95000
data["order_value"][88] = 87000

df = pd.DataFrame(data)

print(df.shape)
df.head()
```

**Task 1 — Inspect & Handle Missing Values**

**1. Print Shape**

```python
print("Shape of DataFrame:")
print(df.shape)
```

**2. Display Data Types**

```python
print("\nData Types:")
print(df.dtypes)
```

**3. Count Missing Values**

```python
print("\nMissing Values Count:")
missing_count = df.isnull().sum()
print(missing_count)
```

**4. Percentage of Missing Values**

```python
print("\nPercentage of Missing Values:")
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage.round(2))
```

**Task 2 — Summarize & Visualize**

**1. Summary Statistics**

```python
summary = df.describe()
print(summary)
```

**Markdown Observations (Example)**

**Observation 1:**
The order_value column has the highest maximum value, which is extremely large compared to the 75th percentile, indicating the presence of outliers.

**Observation 2:**
The delivery_days column shows a relatively consistent distribution, with values mostly between 1 and 14 days.

**2. Histogram for order_value**

```python
plt.figure()
plt.hist(df["order_value"].dropna())

plt.title("Distribution of Order Value")
plt.xlabel("Order Value")
plt.ylabel("Frequency")

plt.show()
```

**3. Box Plot for delivery_days**

```python
plt.figure()
plt.boxplot(df["delivery_days"].dropna())

plt.title("Box Plot of Delivery Days")
plt.ylabel("Delivery Days")

plt.show()
```

**Task 3 — Correlation Analysis**

**1. Compute Correlation Matrix**

```python
correlation_matrix = df.corr(numeric_only=True)
print(correlation_matrix)
```

**2. Plot Heatmap**

```python
plt.figure()
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

**Markdown Observation** (Example)

**Observation:**
There is generally a weak correlation between order_value and delivery_days, indicating that higher order amounts do not strongly affect delivery time.
