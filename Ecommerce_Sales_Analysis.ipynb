**1.Import Libraries and Create Dataset**

```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
```

**Task 1: Basic Grouping and Single Aggregations**


**1.Total Sales by Region**

groupby() - region and sales_amount two columns calculate the sum values.
region_sales.sort_values by - sales_amount column it will ascending on false satatement.(with out ascending)
display the new region_sales value.

```python
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()
region_sales = region_sales.sort_values(by='sales_amount', ascending=False)
print(region_sales)
```

**2. Number of Transactions per Category**

Data frame in groupby - category and transaction_id to find the count values.


```python
category_transaction = df.groupby('category')['transaction_id'].count().reset_index()
print(category_transaction)
```

**3. Average Sales Amount per Salesperson**

Dataframe groupby - salesperson and sales_amount to find the mean values.

```python
avg_salesperson_sales= df.groupby('salesperson')['sales_amount'].mean().reset_index()
print(avg_salesperson_sales)
```

**Task 2: Multi-Column Grouping and Multiple Aggregations**

**1.Total Sales by Region and Category**

Multi column have a groupby region,category,sales_amount to find the sum values.


```python
region_category_sales = df.groupby(['region','category'])['sales_amount'].sum().reset_index()
print(region_category_sales)
```

**2. Salesperson Performance Metrics**

Dataframe groupby salesperson,sales_amount and aggerication(sum,mean,max,min) this all field put together.
So,given output received from the values are sorting and sum in non ascending field.
Display the output value of saleperson_performance.

```python
saleperson_performance=df.groupby('salesperson')['sales_amount'].agg(['sum','mean','max','min']).reset_index()
saleperson_performance = saleperson_performance.sort_values(by='sum', ascending=False)
print(saleperson_performance)
```

**3. Category with Maximum Total Revenue:**

Dataframe groupby category and sales_amount to find the sum values
To given the values to find the idmax vlaues.
Display the output

```python
category_sales = df.groupby('category')['sales_amount'].sum()
top_category = category_sales.idxmax()
print("category with highesht revenue:",top_category)
```

**Task 3: Custom Aggregation and Complete Sales Report**

**1. Define Custom Sales Range Function**

**2. Apply Custom Aggregation by Region**

Dataframe have groupby the region and sales_amount, aggrecation of sum,mean,min,max,sales_range to find the process

Display the output values.

```python
def sales_range(x):
    return x.max() - x.min()

region_analysis = df.groupby('region')['sales_amount'].agg(['sum','mean','min','max', sales_range]
                                                           ).reset_index()
print(region_analysis)
```

**3. Final Sales Summary Report**

Dataframe groupby region,aggrecation of sales_amount to find the sum,mean and customer id field count to check the process.
Display the final report output

```python
final_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index
print(final_report)
```
