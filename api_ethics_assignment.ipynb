**Task 1 — Classify and Handle PII Fields **

| Field               | Classification                       | Action Before Sharing                        | Justification                                                                                                                                 |
| ------------------- | ------------------------------------ | -------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------------------------- |
| **full_name**       | Direct PII                           | **Drop**                                     | A full name uniquely identifies an individual and is not necessary for research analysis. Removing it reduces re-identification risk.         |
| **email**           | Direct PII                           | **Drop**                                     | Email addresses directly identify individuals and are sensitive contact information. They should not be shared with external partners.        |
| **date_of_birth**   | Indirect PII                         | **Mask** (e.g., keep only year or age range) | Full birth date can be used to re-identify individuals. Keeping only age or birth year preserves analytical value while improving privacy.    |
| **zip_code**        | Indirect PII                         | **Mask** (e.g., first 3 digits only)         | ZIP codes can identify small geographic populations. Partial masking maintains regional analysis without exposing precise location.           |
| **job_title**       | Indirect PII                         | **Keep** or **Generalize**                   | Job titles are useful for demographic analysis. If highly specific, they can be generalized (e.g., "Healthcare Worker").                      |
| **diagnosis_notes** | Indirect PII (Sensitive Health Data) | **Pseudonymize**                             | Medical notes may contain identifiers. Replacing patient identifiers with pseudonyms preserves research usefulness while protecting identity. |


**Summary of Privacy Strategy**

**Drop**: Direct identifiers that are not needed

**Mask**: Reduce precision of quasi-identifiers

**Pseudonymize**: Replace identifiers with unique codes

**Generalize**: Convert specific values into broader categories

**This approach aligns with common privacy frameworks such as:**

Data minimization

De-identification

Responsible data sharing
