```python
# Task 1 — Fetch and Validate the Response
import requests

import pandas as pd

# API endpoint
url = "https://restcountries.com/v2/all"

try:
    response = requests.get(url)

    # Print status code
    print("HTTP Status Code:", response.status_code)

    # Check if request was successful
    if response.status_code == 200:
        data = response.json()
        print("Total number of countries retrieved:", len(data))
    else:
        print("Request failed. Please check the API endpoint.")

except requests.exceptions.RequestException as e:
    print("Error occurred while making the request:", e)
```

```python
# Task 2 — Extract and Display Key Fields

# Ensure 'data' was successfully populated from the API call
if 'data' in locals() and data is not None:
    # Create empty list to store extracted data
    countries_list = []

    # Loop through JSON response
    for country in data:
        # Adjusted parsing for restcountries.com v3.1 structure
        common_name = country.get("name", {}).get("common", "Unknown")
        population = country.get("population", 0)
        region = country.get("region", "Unknown")
        area = country.get("area", 0)

        countries_list.append({
            "common_name": common_name,
            "population": population,
            "region": region,
            "area": area
        })

    # Convert list to DataFrame
    df = pd.DataFrame(countries_list)

    # Display first 10 rows
    print("DataFrame head after extracting key fields:")
    display(df.head(10))
else:
    print("Error: 'data' was not populated from the API call. Please check the previous cell's output for issues.")
    df = pd.DataFrame() # Initialize an empty DataFrame to avoid further errors
```

```python
import requests
import pandas as pd

# URL of the CSV file with country data
csv_url = "https://query.data.world/s/q5hthj5t7x547v6n6x7w7k7t7s5t6b"
csv_file_path = "/tmp/countries.csv"

try:
    print(f"Attempting to download CSV from: {csv_url}")
    response = requests.get(csv_url)
    response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)

    with open(csv_file_path, "wb") as f:
        f.write(response.content)
    print(f"CSV downloaded successfully to {csv_file_path}")

    # Load the CSV into a DataFrame
    df_countries = pd.read_csv(csv_file_path)
    print(f"Loaded {len(df_countries)} countries from CSV.")

except requests.exceptions.RequestException as e:
    print(f"Error downloading or loading CSV: {e}")
    df_countries = pd.DataFrame() # Create an empty DataFrame in case of error
```

```python
# Task 2 — Extract and Display Key Fields

# Ensure df_countries is available from the previous step
if 'df_countries' in locals() and not df_countries.empty:
    # Rename columns to match the desired output keys
    df = df_countries.rename(columns={
        "Name": "common_name",
        "Population": "population",
        "Region": "region",
        "Area": "area"
    })

    # Select only the relevant columns and display first 10 rows
    df = df[["common_name", "population", "region", "area"]]
    print("Processed DataFrame head:")
    print(df.head(10))
else:
    print("Error: 'df_countries' not found or is empty. Please ensure the CSV was loaded correctly.")
    df = pd.DataFrame() # Initialize an empty DataFrame to avoid further errors
```

```python
import requests
import pandas as pd

# API endpoint for restcountries v3.1
# The API now requires specifying fields and limits to 10 fields
url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

# Headers to potentially bypass common blocking issues (e.g., User-Agent)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

try:
    print(f"Attempting to fetch data from: {url}")
    response = requests.get(url, headers=headers)

    # Print status code
    print("HTTP Status Code:", response.status_code)

    # Check if request was successful
    if response.status_code == 200:
        data = response.json()
        print("Total number of countries retrieved:", len(data))
        print("Successfully retrieved data from restcountries.com v3.1.")
    else:
        print(f"Request failed with status code {response.status_code}.")
        print("Response content (if any):", response.text) # Print response content for debugging
        data = None # Ensure data is not defined on failure

except requests.exceptions.RequestException as e:
    print("Error occurred while making the request:", e)
    data = None
```
