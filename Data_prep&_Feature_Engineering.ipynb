```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

# --- Dataset Setup ---
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}
df = pd.DataFrame(data)

# --- Task 1: Categorical Encoding ---

# 1. Gender -> Label Encoding
le_gender = LabelEncoder()
df['Gender'] = le_gender.fit_transform(df['Gender'])

# 2. City -> One-Hot Encoding (Drop first to avoid dummy variable trap)
# We use pd.get_dummies for convenience, which is standard in data analysis
df = pd.get_dummies(df, columns=['City'], drop_first=True)

# 3. Size -> Label Encoding with specific order (Ordinal Encoding)
# Since Size has a logical order, we manually map it to ensure Small=0, Medium=1, Large=2
size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)

# --- Task 2: Feature Scaling ---

# We use StandardScaler for Age and Fare.
# Justification: While Fare has an outlier (450), StandardScaler is the industry standard
# for Gradient Boosting and Linear models. If outliers are extreme, RobustScaler
# would be preferred, but for this small dataset, StandardScaler ensures
# a mean of 0 and variance of 1 for stable model convergence.

scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

# Display Final Result
print("Processed DataFrame:")
print(df)
```

Step-by-Step Explanation
1. Categorical Encoding Logic

    Label Encoding (Gender): Used for binary categories (Male/Female). It converts them to 0 and 1.

    One-Hot Encoding (City): Since there is no "rank" between Mumbai, Delhi, and Chennai, we create separate columns. We drop one column (e.g., City_Chennai) because if a passenger isn't from Delhi or Mumbai, they must be from Chennai—this prevents redundant data (the Dummy Variable Trap).

    Ordinal Mapping (Size): Unlike Gender, "Size" has an inherent order. We use a dictionary map to ensure the mathematical distance between Small (0) and Medium (1) is recognized by the model.

2. Feature Scaling Choice

In your task, the Fare jump from 15 to 450 is significant.

    StandardScaler: Transforms data to have a mean of 0 and a standard deviation of 1. It is sensitive to outliers, but it is the most common requirement for many ML algorithms.

    Alternative (RobustScaler): If you wanted the scaling to be completely unaffected by that 450 value, you would use RobustScaler, which scales based on the Interquartile Range (IQR).
