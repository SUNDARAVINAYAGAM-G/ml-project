**Task 3: Handle Missing Values**
age: Fill with median using fillna(df['age'].median())


**1.Age :**
Handle Missing values  fill with age is median method.Median is the middle value in sorted dataset.Four Null values have filled.

```python
df['age'].fillna(df['age'].median(), inplace=True)
df
```

2.**Fill Weight with Median**

Handle Missing values  fill with weight is median method.Median is the middle value in sorted dataset.Five Null values have filled.

```python
df['weight'].fillna(df['weight'].median(), inplace=True)
df
```

3.**Fill Blood Pressure with Median**

Handle Missing values fill with blood_presure is median method.Median is the middle value in sorted dataset.Six Null values have filled.

```python
df['blood_pressure'].fillna(df['blood_pressure'].median(), inplace=True)
df
```

4.**Fill Medication with Mode**

Handle Missing values fill with medication is mode method.Mode is the mode with index value what are assign on that result are append of all null area to be replaced.(mode(0))-Aspirin.Six Null values have filled.


```python
df['medication'].fillna(df['medication'].mode()[0], inplace=True)
df
```

5.**Fill insurance Provider with "Unknown"**

Already we have insurance_provider data type is change categories.so first insurance_provider added categories and

2.**Perform ALL cleaning steps in order: types → missing values → duplicates**

**Step 1 - Convert Data Types**
age,Weight,insurance_provider all data types converted here

```python
df_clean['age']=pd.to_numeric(df_clean['age'], errors='coerce')
df_clean['weight']=pd.to_numeric(df_clean['weight'], errors='coerce')
df_clean['insurance_provider']=df_clean['insurance_provider'].astype('category')
print("Data types after cleaning:")
print(df_clean.dtypes)
```

**Step 2 - Handle Missing Values**

Handle missing value to check and clean the process here


```python
df_clean['age'].fillna(df_clean['age'].median(), inplace=True)
df_clean['weight'].fillna(df_clean['weight'].median(), inplace=True)
df_clean['blood_pressure'].fillna(df_clean['blood_pressure'].median(), inplace=True)
df_clean['medication'].fillna(df_clean['medication'].mode()[0], inplace=True)

# Add 'Unknown' as a category before filling NaN values
df_clean['insurance_provider'] = (df_clean['insurance_provider'].cat.add_categories('Unknown'))
df_clean['insurance_provider'].fillna('Unknown', inplace=True)
```
